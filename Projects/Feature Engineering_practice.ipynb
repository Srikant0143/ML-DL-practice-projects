### Linear DIscriminant Analysis

```python
import sklearn
from sklearn.datasets import load_digits
```

```python
digit = load_digits()
```

```python
X = digit.data
y = digit.target
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state= 1)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as lda
```

```python
sk_lda = lda()
```

```python
sk_lda.fit(X_train, y_train)
```

```python
X_train = sk_lda.transform(X_train)
```

```python
X_test = sk_lda.transform(X_test)
```

```python
y_train
```

```python
print(X_test.shape)
print(X_train.shape)
```

```python
LogReg = LogisticRegression(penalty='l2', solver='liblinear')
```

```python
LogReg.fit(X_train, y_train)
```

```python
y_pred = LogReg.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(y_test, y_pred)
```

#### Confirmatory Factor analysis

```python
from factor_analyzer import (ConfirmatoryFactorAnalyzer, ModelSpecificationParser)
```

```python
model_dict = {"F1": ['radius_mean','perimeter_mean','area_mean','radius_worst','perimeter_worst','diagnosis'], 
              "F2": ['area_worst','concavity_mean','concave points_mean','concavity_worst','concave points_worst']}
```

```python
model_specs = ModelSpecificationParser.parse_model_specification_from_dict(df_corr, model_dict)
```

```python
cfal = ConfirmatoryFactorAnalyzer(model_specs, disp= False)
```

```python
cfal.fit(df_corr.values)
```
